```python
import os
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from datetime import timedelta
from PIL import Image

import scipy.misc

from keras.preprocessing.image import load_img, img_to_array, array_to_img
import tensorflow as tf
```

```python
adidas_final_df = pd.read_csv('adidas_final_df.csv')
img_id = adidas_final_df['id']

img_list = []
for img in img_id:
    img = img + '.jpg'
    img_list.append(img)
```

```python
# 画像表示
plt.figure(figsize=(20,20))
for i in range(100):
    plt.subplot(10,10,i+1)
    path = img_path + '/' + img_list[i]
    img = Image.open(path)
    im_list = np.asarray(img)
    plt.imshow(im_list)
    plt.axis('off')
    plt.title(str(img_list[i]),fontsize=14)
#     plt.savefig('./cifar10.png', bbox_inches='tight')
plt.show()
```

```python
# hw = {'height': 16, 'width': 16}
# hw = {'height': 64, 'width': 64}
hw = {'height': 128, 'width': 128}
# hw = {'height': 256, 'width': 256}


arrlist = []

for img_name in img_list:
    path = img_path + '/' + img_name
    img = load_img(path, target_size = (hw['height'], hw['width']))
    array = img_to_array(img) / 255
    arrlist.append(array)
```

```python
# img_features16
# img_features64
# img_features128

nplist = np.array(arrlist)
np.save('img_features128', nplist)
```
